# 6)

#### Create a spending flag for each user based on the avg price across all their orders based on the criteria:

#### - If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
#### - If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
# We must create a new column called "avg_price_spent"
# This column will contain the mean(avg) of the prices each user spends
ords_prods_merge['avg_price_spent'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Checking the new column
ords_prods_merge.head(50)
```

#### We can see that each user_id has the same value for their avg_price_spent and it changes for each different user

```python
# Now, we'll create another column that labels them as low spender
ords_prods_merge.loc[ords_prods_merge['avg_price_spent'] < 10, 'spending_flag'] = 'Low spender'
```

```python
# Now, we'll create another column that labels them as high spender
ords_prods_merge.loc[ords_prods_merge['avg_price_spent'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# Checking our work
ords_prods_merge.head(50)
```

```python
# Check how many of each there are
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

 Lot's of low spenders!

# 7)

####  Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:
#### If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
#### If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
#### If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
# Creating a new column for median in days_since_prior_order
ords_prods_merge['median_since_prior_orders'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Checking our work
ords_prods_merge.head(50)
```

#### Creating the 3 order frequency flags

```python
ords_prods_merge.loc[ords_prods_merge['median_since_prior_orders'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_since_prior_orders'] > 10) & (ords_prods_merge['median_since_prior_orders'] <= 20), 'order_frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_since_prior_orders'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
# Checking our work
ords_prods_merge.head(50)
```

```python
# Checking how many there are of each customer type
ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```
